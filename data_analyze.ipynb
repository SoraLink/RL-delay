```python
DATA_PATH_PREFIX = 'E:\\Papers\\RL-delay\\Data\\uniform\\'
DATA_PATH_SUFFIX = '\\progress.csv'
```

```python
def read_csv(delay):
    print(DATA_PATH_PREFIX+str(delay)+DATA_PATH_SUFFIX)
    df = pd.read_csv(DATA_PATH_PREFIX+str(delay)+DATA_PATH_SUFFIX)
    y = []
    for p in df['performance']:
        y.append(float(p))
    return df['misc/nupdates'], y
```

```python
def draw_graph():
    _, ax = plt.subplots(3, 3, sharey=True, figsize=(15,15))
    for i in range(3):
        for j in range(3):
            if i==0 and j==0:
                continue 
            x, y = read_csv(i*3+j)
            ax1 = ax[i,j]
            ax1.plot(x, y)
            ax1.set_title(f'delay={i*3+j}')
```

```python
draw_graph()
```
